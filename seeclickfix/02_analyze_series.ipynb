```python
# SET YOUR OWN FREQUENCY FOR TIME SERIES ANALYSIS
# try 'W-SAT' for Sun-Sat, 'M' for month end, 'D' for calendar days, etc
# full options at http://pandas.pydata.org/pandas-docs/stable/timeseries.html#anchored-offsets

config = {
    'frequency': 'W-SAT'
}
```

```python
# RESAMPLE DATASET BY CONFIGURABLE FREQUENCY

# pluck out just the columns we care about
filtered_df = scf_df[['created_at_datetime', 'request_type_title', 'created_at', 'days_create_to_close']]
filtered_df.columns = ['time', 'issue_type', 'total_issues_created', 'med_days_to_close']

# resample the dateset by time, group by request type and aggregate/do math on other cols
filtered_df.set_index('time').groupby('issue_type').resample(config['frequency']).agg({'total_issues_created': np.count_nonzero, 'med_days_to_close': np.median})
```
